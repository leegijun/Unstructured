## 3.3.2 TDM 만들기

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=1000, stop_words='english')
```

```python
tdm = cv.fit_transform(df['reviewText'])
```

### TDM의 기본 형태 

```python
tdm.shape
```

```python
cv.get_feature_names()
```

```python
words = cv.get_feature_names()
```

### 값 보기

```python
tdm[0]
```

```python
doc = tdm[0].toarray()
doc
```

```python
[w for w,c in zip(words, doc.flat) if c >0]
```

```python
used = []
for w,c in zip(words, doc.flat):
    if c>0:
        used.append(w)
        
used
```

```python
[w                          # (4) 단어들의 리스트
 for w, c                   # (2) 각 단어와 빈도 중에
 in zip(words, doc.flat)    # (1) 단어들과 빈도를 짝지은다.
 if c > 0]                  # (3) 빈도가 0보다 큰
```

### 단어 빈도

```python
count = tdm.sum(axis=0)
```

### 빈도순 정렬

```python
import operator
```

```python
from operator import itemgetter
```

```python
word_order = sorted(
    word_count,
    key=operator.itemgetter(1),
    reverse=True)
```
